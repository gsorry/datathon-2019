# Raiffeisen - Merge External Data

```python
import pandas as pd

pd.options.display.max_columns = 400
```

## Read external data

Most of the external data were collected from The Statistical Office of the Republic of Serbia and Open data portal of Republic of Serbia.

- http://www.stat.gov.rs/en-US
- https://data.gov.rs/sr/

### Read cleaned data first

```python
df = pd.read_csv('RFB - Clean Data.csv', index_col=0)
df.shape
```

```python
df.head()
```

### Read and merge Consumer price indices

Consumer price indices measure change of products and services prices that households purchase for satisfying own needs. They are used as inflation measure.

```python
indeksi = pd.read_csv('RFB - Potrosacke Cene Indeksi.csv')
indeksi.shape
```

```python
indeksi.head()
```

### Read and merge Cache Loan labels

We engineered 3 new labels and whether the customer took out a loan in the first 3 months, the first 6 months or did not take out a loan in the first 6 months at all.

```python
zajam = pd.read_csv('RFB - Prvi Zajam.csv')
zajam.shape
```

```python
zajam.head()
```

### Read structure data

```python
imovina = pd.read_csv('RFB - Strukturna Imovina.csv')
imovina.grad = imovina.grad.str.lower()
imovina.shape
```

```python
imovina.head()
```

## Merge loaded data

```python
df = df.merge(indeksi, left_on='month', right_on='Godinamesec')
df[['CUSTOMER_RK', 'CITY', 'month', 'Godinamesec', 'cpi_consumer', 'cpi_excluding']].head()
```

```python
df = df.merge(zajam, left_on='CUSTOMER_RK', right_on='CUSTOMER_RK')
df[['CUSTOMER_RK', 'CITY', 'month', 'Godinamesec', 'cpi_consumer', 'cpi_excluding', 'cash_loan']].head()
```

```python
df = df.merge(imovina, left_on='CITY', right_on='grad')
df[['Godinamesec', 'cpi_consumer', 'cpi_excluding', 'cash_loan', 'CITY', 'grad', 'F', 'P', 'R', 'broj_prijava', 'broj_obveznika']].head()
```

## Save data for later usage

```python
df.to_csv('RFB - Merged Data.csv')
```

```python
# Create table for missing data analysis
def draw_missing_data_table(df):
    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return missing_data
```

```python
draw_missing_data_table(df)
```
